```python
import h2o
from h2o.automl import H2OAutoML

#Starting h2o
h2o.init()

#load Dataset
data = h2o.import_file("diabetes.csv")

#Features and target
x = data.columns[:-1]
y = "Outcome"

#Train AutoML
aml = H2OAutoML(max_models=5, seed=1)
aml.train(x=x, y=y, training_frame=data)

#Best Model
print(aml.leader)
```
