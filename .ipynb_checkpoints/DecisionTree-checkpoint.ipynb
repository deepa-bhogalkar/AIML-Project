## Reading the Dataset

```python
import pandas as pd
df=pd.read_csv('fruits.csv')
df.head()
```

```python
x=df[["Weight","Size"]]
y=df.Class
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=10,shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy")
dt.fit(xtrain,ytrain)
```

```python
print("Training Accuracy=",dt.score(xtrain,ytrain))
print("TestingAccuracy=",dt.score(xtest,ytest))
```

```python
from sklearn.tree import plot_tree
plot_tree(dt,feature_names=x.columns)
```

```python
plot_tree(dt,feature_names=df.columns,class_names=y)
```
