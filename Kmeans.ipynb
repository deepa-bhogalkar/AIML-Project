```python
import pandas as pd
df = pd.read_csv('Income.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['INCOME'],df['SPEND'])
plt.show()
```

```python
x=df[["INCOME","SPEND"]]
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(x)
```

```python
km.labels_
```

```python
km.cluster_centers_
```

```python
km.inertia_
```

```python
Labels=km.predict(x)
df['Labels']= Labels
```

```python
df.head()
```

```python
xnew=[200,115]
km.predict([xnew])
```

```python
plt.scatter(df["INCOME"],df["SPEND"],c=km.labels_,s=50)
plt.show()
```

## Elbow Method

```python
wcss=[]
for i in range(2,17):
    km=KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)
xa=range(2,17)
plt.plot(xa,wcss)
plt.show()
```

```python
km=KMeans(n_clusters=4)
km.fit(df)
```

```python
from sklearn.metrics import silhouette_score
print(silhouette_score(df,km.labels_))
```
