```python
from sklearn .preprocessing import LabelEncoder
le=LabelEncoder()
df['Class']=le.fit_transform(df.Class)
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
df['Weight']=mm.fit_transform(df[['Weight']])
df['Size']=mm.fit_transform(df[['Size']])
df.head()
```

```python
x=df[["Weight","Size"]]
y=df["Class"]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=5,shuffle=True)
```

## Building Model - Linear SVM

```python
from sklearn.svm import LinearSVC
model=LinearSVC()
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print("Accuracy =", accuracy_score(ytest, ypred))
print("Precision =", precision_score(ytest, ypred))
print("Recall =", recall_score(ytest, ypred))
```

## Quick Evaluation

```python
print("Training Accuracy=",model.score(xtrain,ytrain))
print("Testing Accuracy=",model.score(xtrain,ytrain))
```
