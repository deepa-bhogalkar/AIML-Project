## Aggregate function

```python
print(df1['salary'].mode()[0])
```

```python
print(df1['salary'].value_counts())
```

```python
k = df1['rank'].mode()[0]
```

```python
df1['rank'] = df1['rank'].fillna(k)
```

```python
print(df1.isna().sum())
```

```python
numerical_cols = df1.select_dtypes(include=np.number).columns
print(numerical_cols)
categorical_cols = df1.select_dtypes(exclude=np.number).columns
print(categorical_cols)
```

```python
for col in numerical_cols:
    df1[col] = df1[col].fillna(df1[col].mean)
    print(df1[col])
```

# Imputing the values

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy="mean")
```

```python

df1[numerical_cols] = imp.fit_transform(df1[numerical_cols])
```
